```python
import numpy as np
from astroquery.gaia import Gaia
from astropy.coordinates import SkyCoord
import matplotlib.pyplot as plt
```

```python
koordinater = SkyCoord.from_name('NGC 2362')#Skriv in namnet på er stjärnhop här och se om ni får ut rätt koordinater

koordinater
```

```python
query_M45 = '''
SELECT parallax, bp_rp, phot_g_mean_mag
FROM gaiadr3.gaia_source_lite
WHERE DISTANCE(56.601, 24.114, ra, dec) < 1
AND parallax IS NOT NULL
AND bp_rp IS NOT NULL
AND phot_g_mean_mag IS NOT NULL
AND parallax > 0
AND 1000/parallax > 110
AND 1000/parallax < 150
AND parallax_over_error > 2
'''

job = Gaia.launch_job_async(query_M45)#skickar till databasen

M45 = job.get_results()#Hämtar hem resultaten

M45#visar resultaten
```

```python
M45['abs_mag_g'] = M45['phot_g_mean_mag'] - 5 * np.log10(100/M45['parallax'])
```

```python
query_M4 = '''
SELECT parallax, bp_rp, phot_g_mean_mag
FROM gaiadr3.gaia_source_lite
WHERE DISTANCE(245.89675, -26.52575, ra, dec) < 1
AND parallax IS NOT NULL
AND bp_rp IS NOT NULL
AND phot_g_mean_mag IS NOT NULL
AND parallax > 0
AND 1000/parallax > 1500
AND 1000/parallax < 1900
AND parallax_over_error > 2
'''

job = Gaia.launch_job_async(query_M4)#skickar till databasen

M4 = job.get_results()#Hämtar hem resultaten

M4#visar resultaten
```

```python
M4['abs_mag_g'] = M4['phot_g_mean_mag'] - 5 * np.log10(100/M4['parallax'])
```

```python
query_M67 = '''
SELECT parallax, bp_rp, phot_g_mean_mag
FROM gaiadr3.gaia_source_lite
WHERE DISTANCE(132.846, 11.814, ra, dec) < 0.5
AND parallax IS NOT NULL
AND bp_rp IS NOT NULL
AND phot_g_mean_mag IS NOT NULL
AND parallax > 0
AND 1000/parallax > 700
AND 1000/parallax < 1000
AND parallax_over_error > 2
'''

job = Gaia.launch_job_async(query_M67)#skickar till databasen

M67 = job.get_results()#Hämtar hem resultaten

M67
```

```python
M67['abs_mag_g'] = M67['phot_g_mean_mag'] - 5 * np.log10(100/M67['parallax'])
```

```python
from matplotlib import colors

temp_M4 = M4['bp_rp']
magnitud_M4 = M4['abs_mag_g']

temp_M45 = M45['bp_rp']
magnitud_M45 = M45['abs_mag_g']

temp_M67 = M67['bp_rp']
magnitud_M67 = M67['abs_mag_g']

fig, ax = plt.subplots()#Skapar ett figur och axel objekt 

ax.scatter(temp_M4, magnitud_M4, alpha=0.1, s=1, color='k', zorder=0, label='M4')#Här plottar vi upp varje stjärna som en punkt
ax.scatter(temp_M45, magnitud_M45, alpha=0.1, s=1, color='r', zorder=0, label='M45')#Här plottar vi upp varje stjärna som en punkt
ax.scatter(temp_M67, magnitud_M67, alpha=0.1, s=1, color='b', zorder=0, label='M67')#Här plottar vi upp varje stjärna som en punkt
ax.invert_yaxis()

ax.set_xlabel('Bp-Rp')#Axel titlar
ax.set_ylabel('Magntiud')
ax.legend(markerscale=10)

plt.show()
```
